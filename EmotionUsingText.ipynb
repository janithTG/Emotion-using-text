```python
from datasets import load_dataset
```

```python
emotions = load_dataset('SetFit/emotion')
```

```python
emotions
```

```python
def tokenize(batch):
  return tokenizer(batch["text"], padding = True, truncation = True)
```

```python
emotions_encoded = emotions.map(tokenize, batched=True, batch_size = None)
```

```python
emotions_encoded
```

```python
import tensorflow as tf
```

```python
# setting 'input_ids', 'attention_mask', 'token_type_ids', and 'label'
# to the tensorflow format. Now if you access this dataset you will get these
# columns in `tf.Tensor` format

emotions_encoded.set_format('tf',
                            columns=['input_ids', 'attention_mask', 'token_type_ids', 'label'])

# setting BATCH_SIZE to 64.
BATCH_SIZE = 64

def order(inp):
    '''
    This function will group all the inputs of BERT
    into a single dictionary and then output it with
    labels.
    '''
    data = list(inp.values())
    return {
        'input_ids': data[1],
        'attention_mask': data[2],
        'token_type_ids': data[3]
    }, data[0]

# converting train split of `emotions_encoded` to tensorflow format
train_dataset = tf.data.Dataset.from_tensor_slices(emotions_encoded['train'][:])
# set batch_size and shuffle
train_dataset = train_dataset.batch(BATCH_SIZE).shuffle(1000)
# map the `order` function
train_dataset = train_dataset.map(order, num_parallel_calls=tf.data.AUTOTUNE)

# ... doing the same for test set ...
test_dataset = tf.data.Dataset.from_tensor_slices(emotions_encoded['test'][:])
test_dataset = test_dataset.batch(BATCH_SIZE)
test_dataset = test_dataset.map(order, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
inp, out = next(iter(train_dataset)) # a batch from train_dataset
print(inp, '\n\n', out)
```

```python
class BERTForClassification(tf.keras.Model):

    def __init__(self, bert_model, num_classes):
        super().__init__()
        self.bert = bert_model
        self.fc = tf.keras.layers.Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.bert(inputs)[1]
        return self.fc(x)
```

```python
classifier = BERTForClassification(model, num_classes = 6)

classifier.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-5),
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics = ['accuracy']
)
```

```python
history = classifier.fit(
    train_dataset,
    epochs = 3
)
```

```python
classifier.evaluate(test_dataset)
```

```python
# Example of how to make predictions on new text
new_text = "i feel loyal to style"

# Step 1: Tokenize the new text
new_text_tokens = tokenizer(new_text, padding=True, truncation=True, return_tensors='tf')

# Step 2: Convert to TensorFlow format
new_text_inputs = {
    'input_ids': new_text_tokens['input_ids'],
    'attention_mask': new_text_tokens['attention_mask'],
    'token_type_ids': new_text_tokens['token_type_ids']
}

# Step 3: Make Predictions
predictions = classifier.predict(new_text_inputs)

# You can now interpret the predictions based on your problem.
# For example, if you have a classification task, you might want to
# get the predicted class labels.
predicted_labels = tf.argmax(predictions, axis=1).numpy()

print("Predicted Labels:", predicted_labels)
# Assuming you have a mapping dictionary like this:
label_mapping = {
    0: 'Sadness',
    1: 'Joy',
    2: 'Love',
    3: 'Anger',
    4: 'Fear',
    5: 'Surprise'

}

# Map class indices to label names
predicted_label_names = [label_mapping[label] for label in predicted_labels]

print("Predicted Label Names:", predicted_label_names)
```
